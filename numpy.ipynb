Integrals and derivative

```python
x = np.linspace(1,10,100)
y = 1/x**2 * np.sin(x)
# derivatvie 
dydx = np.gradient(y, x)
# integral
y_int = np.cumsum(y)*(10/100) #an integral is just the sum over the inveral
fig, ax = plt.subplots()
ax.plot(x,y, label="f(x)");
ax.plot(x,dydx, label="dy/dx")
ax.plot(x,y_int)
plt.show()
```

Consider the flower petal $r(\theta) = 1 + \frac{3}{4}sin(3\theta)$ for $0\leq \theta \leq 2\pi$

1. Make a plot of the flower (finx x and y points)
2. Compute the area using the calcululs formula $A = \int_0^{2\pi}\frac{1}{2}r^2d\theta$
3. Compute the arclenght using the calculus formula $L = \int_0^{2\pi}\sqrt{r^2+\left(\frac{dr}{d\theta}\right)^2}d\theta$

```python
#1
theta = np.linspace(0,2*np.pi, 1000)
def r(theta):
    return 1+3/4*np.sin(3*theta)

x = r(theta)*np.cos(theta)
y = r(theta)*np.sin(theta)
plt.plot(x,y);
```

```python
#2
```

```python
A = 1/2*sum(r(theta)**2)*(theta[1]-theta[0])
np.round(A,2)
```

# Question 4
The power emitted by a blackbody is $P = A\sigma\varepsilon T^4$. After measuring the temperature of a star you find that $T(t) = T_0 \frac{1}{1+e^{-kt}}$. Plot the total energy emitted by the star as a function of time using the that $E(t) = \int_{t'=0}^{t'=t} P(t')dt'$

Now to solve this and plot this, we need to know the constant. So the idea is to plot what we know and rearange the terms:
$$ \frac{P(t)}{A\sigma\varepsilon T_0^4} = \left(\frac{1}{1+e^{-kt}}\right)^4$$ 

Now we can express everything in terms of our Power with respect to $kt$ since we do not know $k$

```python
kt = np.linspace(0,3, 100)
P_no_dim = (1/(1+np.exp(-kt)))**4
E = np.cumsum(P_no_dim)*(kt[1]-kt[0])
```

```python
# plot
plt.plot(kt, P_no_dim, label="Power")
plt.plot(kt, E, label="Energy")
plt.legend()
plt.xlabel("kt")
plt.ylabel(r"$\frac{P(t)}{A\sigma\varepsilon T_0^4}$", fontsize=15)
plt.show()
```
